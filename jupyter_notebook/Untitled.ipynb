```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('../data/menu_data.xlsx')
```

```python
df.head()
```

```python
X = df[['메뉴', '빵', '치즈']].to_numpy()
```

```python
cnt = {}
```

```python
for i in range(len(X)):
    if not(str(X[i]) in cnt.keys()):
        cnt.update({str(X[i]):1})
    else:
        cnt[str(X[i])] += 1        
```

```python
cnt
```

```python
sorted_cnt = sorted(cnt.items(), key = lambda item: item[1])
```

```python
sorted_cnt
```

```python
val = [i[0] for i in sorted_cnt]
```

```python
cnt2 = [i[1] for i in sorted_cnt]
```

```python
y = [i for i in range(len(cnt))]
plt.figure(figsize=(10,80))
plt.barh(y, cnt2)
plt.yticks(y, val)
plt.show()
```

```python
cnt_series = pd.Series(cnt2)
```

```python
pd.DataFrame(cnt_series.describe())
```
