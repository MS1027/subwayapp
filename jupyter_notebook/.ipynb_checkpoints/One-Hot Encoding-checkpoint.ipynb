# 원-핫 인코딩

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('../data/menu_data.xlsx')
```

```python
df.head()
```

```python
menu_dummies = pd.get_dummies(df['메뉴'])
```

```python
bread_dummies = pd.get_dummies(df['빵'])
```

```python
cheese_dummies = pd.get_dummies(df['치즈'])
```

```python
menu_dummies.columns = ['쉬림프', '로티세리 바비큐 치킨', '이탈리안 비엠티', '로스트 치킨', '에그마요', 'K-바비큐', '풀드 포크 바비큐',
                       '비엘티', '햄', '참치', '써브웨이 클럽', '베지', '스테이크 & 치즈', '스파이시 이탈리안', '치킨 데리야끼']
```

```python
bread_dummies.columns = ['허니오트', '하티', '위트', '파마산 오레가노', '화이트', '플랫브래드']
```

```python
cheese_dummies.columns = ['아메리칸 치즈', '슈레드 치즈', '모차렐라 치즈', '치즈 제외']
```

```python
df_tmp1 = pd.merge(menu_dummies, bread_dummies, left_index=True, right_index=True)
```

```python
df_tmp2 = pd.merge(df_tmp1, cheese_dummies, left_index=True, right_index=True)
```

```python
df_tmp2.shape
```

```python
df_tmp2.head()
```

```python
df2 = pd.read_excel('../data/input_data.xlsx')
```

```python
df2.iloc[:, 18:]
```

```python
df_input = pd.merge(df_tmp2, df2.iloc[:, 18:], left_index=True, right_index=True)
```

```python
df_input.head()
```

```python
df_input.to_excel('../data/One-Hot.xlsx', index=False)
```
