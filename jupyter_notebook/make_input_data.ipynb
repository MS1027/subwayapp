```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
```

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
```

```python
from sklearn.metrics.pairwise import linear_kernel
```

```python
tfidf_matrix = tfidf.fit_transform(des['description'])
```

```python
tfidf_matrix = tfidf_matrix.toarray()
```

```python
tfidf_matrix = tfidf_matrix.tolist()
```

```python
df_des = []
```

```python
for i in range(df.shape[0]):
    df_des.append(tfidf_matrix[df.loc[i, '메뉴']-1])
```

```python
df_tmp = pd.DataFrame(df_des)
```

```python
df_input = pd.merge(df, df_tmp, left_index=True, right_index=True)
```
